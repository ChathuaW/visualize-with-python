# Data Visualisation Using Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import warnings
warnings.simplefilter("ignore",UserWarning)
```

```python
housing = data("Housing")
```

```python
housing.head()
```

```python
housing.plot();
```

```python
housing.plot(
    y = "price"
);
```

```python
housing.plot(
    y = "price",
    xlabel = "House Index",
    ylabel = "Prices",
    title = "House Pricing"
);
```

```python
housing.plot(
    y = "price",
    xlabel = "Housing index",
    ylabel = "Price",
    title = "Housing Price",
    legend = False
);
```

```python
housing.plot(
    y = "price",
    xlabel = "Housing Index",
    ylabel = "Price",
    title = "Housing Price",
    legend = False,
    figsize = (20,10)
);
```

```python
housing.plot(
    y = ["price","lotsize"],
    xlabel = "House Index",
    ylabel = "Price",
    title = "House Prices Vs House Price & Lotsize",
    subplots = True
);
```

```python
housing.plot(
    y = ["price","lotsize"],
    xlabel = "Housing Index",
    ylabel = "Price",
    subplots = True,
    layout = (1,2)
);
```

```python
housing.plot(
    y = ["price","lotsize"],
    xlabel = "Housing Index",
    ylabel = "Price",
    subplots = True,
    layout = (1,2)
),
plt.tight_layout()
```
