# Data Visualisation Using Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
import warnings
warnings.simplefilter('ignore',UserWarning)
```

```python
iris_df = data('iris')
```

```python
iris_df.head()
```

```python
sns.jointplot(data = iris_df,x = "Sepal.Length", y = "Petal.Width", hue = "Species");
```

```python
housing = data("Housing")
```

```python
sns.relplot(data = housing, x = "lotsize", y = "price",height = 5,aspect = 2);
```

```python
sns.relplot(data = housing, x = "lotsize", y ="price", height = 5, aspect = 2);
plt.xlabel("Lot Size")
plt.ylabel("Price")
plt.title("Lot Size Vs Price")
```
